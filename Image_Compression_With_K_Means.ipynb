```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from sklearn.cluster import KMeans
```

```python
Img=Image.open("/content/Screenshot 2025-10-10 221817.png")
img=np.array(Img)
```

```python
plt.imshow(img)
plt.axis("off")
plt.show()
```

```python
h,w,ch=img.shape
```

```python
h
```

```python
w
```

```python
img=img.reshape([-1,3])
```

```python
K=64
```

```python
model=KMeans(n_clusters=K)
```

```python
model.fit(img)
```

```python
c=model.cluster_centers_
```

```python
labels=model.labels_
```

```python
labels.shape
```

```python
new_img=c[labels].astype(np.uint8)
```

```python
new_img=new_img.reshape(h,w,ch)
```

```python
new_img.shape
```

```python
plt.imshow(new_img)
plt.axis("off")
plt.show()
```

```python
Image.fromarray(new_img).save("New_img_compressed.png")
```
